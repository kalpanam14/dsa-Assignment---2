```python
#Delete the elements in an linked list whose sum is equal to zero
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def delete_zero_sum_nodes(head):
    dummy = Node(0)
    dummy.next = head
    prefix_sum = 0
    prefix_sum_map = {}
    current = dummy.next

    while current:
        prefix_sum += current.data

        if prefix_sum in prefix_sum_map:
            prev = prefix_sum_map[prefix_sum]
            prev.next = current.next
            current = current.next
            continue
        else:
            prefix_sum_map[prefix_sum] = current

        current = current.next

    return dummy.next

# Example usage
# Create a linked list: 3 -> 4 -> -7 -> 5 -> -6 -> 6 -> -2 -> 1
head = Node(3)
node1 = Node(4)
node2 = Node(-7)
node3 = Node(5)
node4 = Node(-6)
node5 = Node(6)
node6 = Node(-2)
node7 = Node(1)

head.next = node1
node1.next = node2
node2.next = node3
node3.next = node4
node4.next = node5
node5.next = node6
node6.next = node7

# Delete zero sum nodes
new_head = delete_zero_sum_nodes(head)

# Print the resulting linked list
current = new_head
while current:
    print(current.data, end=" -> ")
    current = current.next
print("None")
```

```python
#Reverse a linked list in groups of given size
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def reverse_in_groups(head, k):
    if not head or k == 0 or k == 1:
        return head

    # Function to reverse a linked list
    def reverse_list(node):
        prev = None
        current = node
        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        return prev

    dummy = Node(0)
    dummy.next = head
    prev_group_tail = dummy

    while True:
        current = prev_group_tail.next
        group_head = current

        # Find the end of the current group
        for _ in range(k - 1):
            if current:
                current = current.next
            else:
                break

        if not current:
            break

        next_group_head = current.next
        current.next = None

        # Reverse the current group
        reversed_group_head = reverse_list(group_head)
        prev_group_tail.next = reversed_group_head
        group_head.next = next_group_head

        prev_group_tail = group_head

    return dummy.next

# Example usage
# Create a linked list: 1 -> 2 -> 3 -> 4 -> 5 -> 6 -> 7 -> 8 -> None
head = Node(1)
node1 = Node(2)
node2 = Node(3)
node3 = Node(4)
node4 = Node(5)
node5 = Node(6)
node6 = Node(7)
node7 = Node(8)

head.next = node1
node1.next = node2
node2.next = node3
node3.next = node4
node4.next = node5
node5.next = node6
node6.next = node7

# Reverse in groups of size 3
k = 3
new_head = reverse_in_groups(head, k)

# Print the resulting linked list
current = new_head
while current:
    print(current.data, end=" -> ")
    current = current.next
print("None")
```

```python
#Merge a linked list into another linked list at alternate positions
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def merge_alternate(head1, head2):
    if not head1:
        return head2
    if not head2:
        return head1

    current1 = head1
    current2 = head2

    while current1 and current2:
        next1 = current1.next
        next2 = current2.next

        current1.next = current2
        current2.next = next1

        current1 = next1
        current2 = next2

    return head1

# Example usage
# Create the first linked list: 1 -> 2 -> 3 -> None
head1 = Node(1)
node1 = Node(2)
node2 = Node(3)

head1.next = node1
node1.next = node2

# Create the second linked list: 4 -> 5 -> 6 -> 7 -> None
head2 = Node(4)
node3 = Node(5)
node4 = Node(6)
node5 = Node(7)

head2.next = node3
node3.next = node4
node4.next = node5

# Merge at alternate positions
merged_head = merge_alternate(head1, head2)

# Print the resulting merged linked list
current = merged_head
while current:
    print(current.data, end=" -> ")
    current = current.next
print("None")
```

```python
#In an array, Count Pairs with given sum
def count_pairs_with_sum(arr, target_sum):
    pair_count = 0
    num_count = {}

    for num in arr:
        complement = target_sum - num

        if complement in num_count:
            pair_count += num_count[complement]

        if num in num_count:
            num_count[num] += 1
        else:
            num_count[num] = 1

    return pair_count

# Example usage
arr = [1, 5, 7, -1, 5]
target_sum = 6

pair_count = count_pairs_with_sum(arr, target_sum)
print(f"Number of pairs with sum {target_sum}: {pair_count}")
```

```python
#Find duplicates in an array
def find_duplicates(arr):
    count_dict = {}
    duplicates = []

    for num in arr:
        if num in count_dict:
            count_dict[num] += 1
            if count_dict[num] == 2:
                duplicates.append(num)
        else:
            count_dict[num] = 1

    return duplicates

# Example usage
arr = [1, 2, 3, 4, 2, 5, 6, 3, 6]
duplicates = find_duplicates(arr)
print("Duplicates:", duplicates)
```

```python
#Find the Kth largest and Kth smallest number in an array
def find_kth_largest_smallest(arr, k):
    arr.sort()  # Sort the array in ascending order

    kth_smallest = arr[k - 1]
    kth_largest = arr[-k]

    return kth_largest, kth_smallest

# Example usage
arr = [7, 3, 9, 2, 1, 5, 6]
k = 3

kth_largest, kth_smallest = find_kth_largest_smallest(arr, k)
print(f"Kth Smallest: {kth_smallest}")
print(f"Kth Largest: {kth_largest}")
```

```python
#Move all the negative elements to one side of the array
def move_negatives(arr):
    left = 0
    right = len(arr) - 1

    while left <= right:
        if arr[left] < 0 and arr[right] < 0:
            left += 1
        elif arr[left] >= 0 and arr[right] < 0:
            arr[left], arr[right] = arr[right], arr[left]
            left += 1
            right -= 1
        elif arr[left] >= 0 and arr[right] >= 0:
            right -= 1
        else:
            left += 1
            right -= 1

    return arr

# Example usage
arr = [-1, 2, -3, 4, 5, -6, -7, 8, -9]
result = move_negatives(arr)
print("Array with negatives moved to one side:", result)
```

```python
#Reverse a string using a stack data structure
def reverse_string(string):
    stack = []

    # Push each character of the string onto the stack
    for char in string:
        stack.append(char)

    reversed_string = ""

    # Pop each character from the stack to build the reversed string
    while stack:
        reversed_string += stack.pop()

    return reversed_string

# Example usage
string = "Hello, World!"
reversed_str = reverse_string(string)
print("Reversed String:", reversed_str)
```

```python
Evaluate a postfix expression using stack
def evaluate_postfix(expression):
    stack = []

    for char in expression:
        if char.isdigit():
            stack.append(int(char))
        else:
            operand2 = stack.pop()
            operand1 = stack.pop()

            result = perform_operation(char, operand1, operand2)
            stack.append(result)

    return stack.pop()

def perform_operation(operator, operand1, operand2):
    if operator == '+':
        return operand1 + operand2
    elif operator == '-':
        return operand1 - operand2
    elif operator == '*':
        return operand1 * operand2
    elif operator == '/':
        return operand1 / operand2
    else:
        raise ValueError("Invalid operator")

# Example usage
postfix_expression = "62/3-42*+"
result = evaluate_postfix(postfix_expression)
print("Result:", result)
```

```python
#Implement a queue using the stack data structure
class Queue:
    def __init__(self):
        self.enqueue_stack = []
        self.dequeue_stack = []

    def enqueue(self, item):
        self.enqueue_stack.append(item)

    def dequeue(self):
        if not self.dequeue_stack:
            # Transfer elements from enqueue_stack to dequeue_stack
            while self.enqueue_stack:
                self.dequeue_stack.append(self.enqueue_stack.pop())

        if not self.dequeue_stack:
            # Both stacks are empty, the queue is empty
            return None

        return self.dequeue_stack.pop()

    def is_empty(self):
        return not self.enqueue_stack and not self.dequeue_stack

    def size(self):
        return len(self.enqueue_stack) + len(self.dequeue_stack)

# Example usage
queue = Queue()

queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)

print("Queue size:", queue.size())

print("Dequeued item:", queue.dequeue())
print("Dequeued item:", queue.dequeue())
print("Dequeued item:", queue.dequeue())

print("Is the queue empty?", queue.is_empty())
```
